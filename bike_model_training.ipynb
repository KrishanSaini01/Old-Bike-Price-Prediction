```python
x=df.drop(['price'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

# SVM (SVR)

```python
from sklearn.svm import SVR
sv=SVR()
sv.fit(x_train,y_train)
```

```python
pred=sv.predict(x_test)
print(pred)
print('SVR test score:-',sv.score(x_test,y_test))
print('SVR train score:-',sv.score(x_train,y_train))
```

# SVM (SVC)

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
```

```python
pred=sv.predict(x_test)
print(pred)
print('SVC test score:-',sv.score(x_test,y_test))
print('SVC train score:-',sv.score(x_train,y_train))
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
print(pred)
print('LinearRegression test score:-',lr.score(x_test,y_test))
print('LinearRegression train score:-',lr.score(x_train,y_train))
```

```python
from sklearn.linear_model import LogisticRegression
log =LogisticRegression()
log.fit(x_train,y_train)
```

```python
pred=log.predict(x_test)
print(pred)
print("LogisticRegression training",log.score(x_train,y_train))
print("LogisticRegression testing",log.score(x_test,y_test))
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
pred=dt.predict(x_test)
print(pred)
print('DecisionTreeClassifier test score:-',dt.score(x_test,y_test))
print('DecisionTreeClassifier train score:-',dt.score(x_train,y_train))
```
